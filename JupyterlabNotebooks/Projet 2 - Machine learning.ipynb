#### Plusieurs genres

```python
#Créer un DF filtré SUR LA PERIODE ET LES GENRES renseignés par l'utilisateur

df_year_genre_choisi_bis = pd.DataFrame()

for genre in liste_genres_choisi : 
    df_genre= df_machine_learning[ (df_machine_learning[genre] == True) 
                                  & (df_machine_learning["startYear"] >= start_year) 
                                  & (df_machine_learning["startYear"] <= end_year) ]
    df_year_genre_choisi_bis = pd.concat([df_year_genre_choisi_bis , df_genre])
                                                   

df_year_genre_choisi_bis.head()
```
