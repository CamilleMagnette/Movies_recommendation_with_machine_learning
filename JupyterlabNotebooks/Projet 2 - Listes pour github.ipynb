# Récupération des titres de film

```python
#récupération des données de la colonne "primarytitle" en un DF
primarytitle = df_machine_learning[['primaryTitle', 'tconst']]
```

```python
#Récupération des données de la colonne "originaltitle" en un df
originaltitle = df_machine_learning[["originalTitle", "tconst"]]
```

```python
#Changement de nom de la colonne "originalTitle"
originaltitle = originaltitle.rename(columns = {"originalTitle" : "primaryTitle"})
```

```python
originaltitle.head(10)
```

```python
#Récupération des données de la colonne "French_Title" en un df
French_Title = df_machine_learning[["French_Title", "tconst"]]
```

```python
#Changement de nom de la colonne "French_Title"
French_Title = French_Title.rename(columns = {"French_Title" : "primaryTitle"})
```

```python
French_Title.head(10)
```

```python
#Création d'un df avec les données des deux df créés ci-dessus
liste_films = pd.concat([primarytitle, originaltitle, French_Title])
```

# Récupération des genres 

```python
# Chargement en local du fichier de data viz pour avoir les genres
merge_final_nettoye = pd.read_csv("../DATASETS NETTOYES/merge_final.csv", sep = ",")  
```

```python
#récupération des données de la colonne "genres" en un DF
liste_genres = merge_final_nettoye.groupby(["genres"]).count().reset_index().sort_values(by = "tconst", ascending = False).iloc[:,:2]
liste_genres
```

```python
# get dummies pour séparer les genres
liste_genres = liste_genres["genres"].str.get_dummies(sep=',').astype(bool)
```

```python
liste_genres
```

```python
# récupérer le titre de chaque colonne 
columns = liste_genres.columns
```

```python
columns
```

```python
# On stocke nos genres dans un nouveau dataframe
liste_genres = pd.DataFrame({"genres" : columns})
```

```python
liste_genres
```

## Enregistrer fichier corrigé sur l'ordi en pickle

```python
liste_genres.to_pickle("../FICHIERS POUR MACHINE LEARNING/liste_genres.pkl.gz", compression = "gzip")
```
