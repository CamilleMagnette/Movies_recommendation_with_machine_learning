# METHODE DUMMIES

## Spliter les genres

```python
# encodage + tranformation en boolean
df_genres_2 = merge_final_nettoye_Pkl_2["genres"].str.get_dummies(sep=',').astype(bool) 
df_genres_2
```

```python
# on concat sur les colonnes le df_genres avec le df d'origne sans la colonne "genres"
df_dummies_2 = pd.concat([merge_final_nettoye_Pkl_2.drop(columns="genres"), df_genres_2 ], axis=1)
df_dummies_2
```

## Spliter les Startyear (ne se fait plus)

```python
# encodage + tranformation en boolean  en convertissant en string car get dummies est une methode en string
df_startyear_2 = df_dummies_2["startYear"].astype(str).str.get_dummies(sep=',').astype(bool) 
df_startyear_2
```

```python
# on concat sur les colonnes le df_primaryName avec le df_dummies sans la colonne "primaryName"
df_dummies_2 = pd.concat([df_dummies_2.drop(columns="startYear"), df_startyear_2 ], axis=1)
df_dummies_2
```

## Spliter les runtimeMinutes (ne se fait plus)

```python
df_runtimeminutes_2 = df_dummies_2["runtimeMinutes"].astype(str).str.get_dummies(sep=',').astype(bool) 
df_runtimeminutes_2
```

```python
# on concat sur les colonnes le df_primaryName avec le df_dummies sans la colonne "primaryName"
df_dummies_2 = pd.concat([df_dummies_2.drop(columns="runtimeMinutes"), df_runtimeminutes_2 ], axis=1)
df_dummies_2
```

## Spliter les primaryName

```python
# encodage + tranformation en boolean
df_primaryName_2 = df_dummies_2["primaryName"].str.get_dummies(sep=',').astype(bool) 
df_primaryName_2
```

```python
# on concat sur les colonnes le df_primaryName avec le df_dummies sans la colonne "primaryName"
df_dummies_2 = pd.concat([df_dummies_2.drop(columns="primaryName"), df_primaryName_2 ], axis=1)
df_dummies_2
```

## Enregistrer le nouveau fichier dummies en PICKLE

```python
df_dummies_2.to_pickle("../df_machine_learning.pkl.gz", compression = "gzip")
```
