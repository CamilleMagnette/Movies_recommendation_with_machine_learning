## Spliter les genres : methode dummies

### Fichier dataviz

```python
# encodage
merge_final_nettoye_Pkl["genres"].str.get_dummies(sep=',') 
#En ligne => les films
#En colonne => les genres 
# plusieurs lignes pour un même film
```

```python
# encodage + tranformation en boolean
df_genres = merge_final_nettoye_Pkl["genres"].str.get_dummies(sep=',').astype(bool) 
df_genres
```

```python
# on concat sur les colonnes le df_genres avec le df d'origne sans la colonne "genres"
df_dummies = pd.concat([merge_final_nettoye_Pkl.drop(columns="genres"), df_genres ], axis=1)
df_dummies
```

### Fichier machine learning

```python
# encodage + tranformation en boolean
df_genres_2 = merge_final_nettoye_Pkl_2["genres"].str.get_dummies(sep=',').astype(bool) 
df_genres_2
```

```python
# on concat sur les colonnes le df_genres avec le df d'origne sans la colonne "genres"
df_dummies_2 = pd.concat([merge_final_nettoye_Pkl_2.drop(columns="genres"), df_genres_2 ], axis=1)
df_dummies_2
```
